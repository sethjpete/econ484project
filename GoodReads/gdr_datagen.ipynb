```python
import pandas as pd
import statistics
```

```python
df1 = pd.read_csv('../dylancastillo/books.csv')
df1['isbn'] = df1['isbn10']
print(len(df1))
df1.head(3)
```

```python
df2 = pd.read_csv('books.csv', index_col=0)
print(len(df2))
df2.head(3)
```

```python
# Merged Data Sets
df = pd.merge(df1, df2, on='isbn', how='inner')
df
```

```python
# More data cleaning
def min_length(x, y):
    if len(str(x)) < len(str(y)):
        return x
    else:
        return y
    
df['title'] = df.apply(lambda x: min_length(x['title_x'], x['title_y']), axis=1)
df['authors'] = df.apply(lambda x: min_length(x['authors_x'], x['authors_y']), axis=1)
df['avg_rating'] = df.apply(lambda x: statistics.mean([x['average_rating_x'], x['average_rating_y']]), axis=1)
df['num_pages'] = df.apply(lambda x: statistics.mean([x['num_pages_x'], x['num_pages_y']]), axis=1)
df['num_ratings'] = df.apply(lambda x: statistics.mean([x['ratings_count_x'], x['ratings_count_y']]), axis=1)
subset = ['isbn10', 'title', 'subtitle', 'authors', 'categories', 'description', 'avg_rating', 'num_pages', 'num_ratings', 'publisher', 'publication_date']
df = df[subset]
df.sort_values(by = 'num_ratings', ascending=False, inplace=True)
df
```

```python
import re
def gen_title(x):
    if str(x['subtitle']) == 'nan' or x['subtitle'] == 'NaN' or x['subtitle'] == None:
        return x['title']
    else:
        return x['title'] + ' (' + str(x['subtitle']) + ')'
df['Title'] = df.apply(lambda x: gen_title(x), axis=1)
df['Author'] = df['authors']
df['Subjects'] = df['categories']
df['Subjects'] = df['Subjects'].apply(lambda x: re.sub(r'[^a-zA-Z0-9 ]', '', str(x)).strip())
df['ISBN'] = df['isbn10']
df['Publisher'] = df['publisher']
df['Publication Year'] = df['publication_date'].str[-4:]
```

```python
df = df[["ISBN","Title","Author","Publisher","Publication Year",'Subjects']]
df
```

```python
df.to_csv('../Exported/GRD_Sel.csv', index=False)
```
