```python
import pandas as pd
```

```python
# McCord has several named problem that we have to deal with
mccord = pd.read_excel('McCord.xlsx', sheet_name='Sheet1')['ID']
base_mccord = pd.read_csv('../Exported/HW/nyt2.csv').drop(columns=['Unnamed: 0'])
```

```python
base_mccord['MAPPED_ID'] = mccord
base_mccord = base_mccord.query('MAPPED_ID != "PROBLEM"').reset_index(drop=True)
base_mccord['MAPPED_ID'].fillna(-1, inplace=True)
base_mccord['MAPPED_ID'] = base_mccord['MAPPED_ID'].astype(str).str.split(',', expand=False)
base_mccord = base_mccord.explode('MAPPED_ID')
base_mccord['MAPPED_ID'] = base_mccord['MAPPED_ID'].astype(int)
base_mccord.set_index('MAPPED_ID', inplace=True)
base_mccord.sort_index(ascending=False)
# We're good!
```

```python
isaac = pd.read_excel('Isaac.xlsx', sheet_name='Sheet1')['ID']
base_isaac = pd.read_csv('../Exported/HW/nyt1.csv').drop(columns=['Unnamed: 0'])
```

```python
base_isaac['MAPPED_ID'] = isaac
base_isaac = base_isaac.query('MAPPED_ID != "PROBLEM"').reset_index(drop=True)
base_isaac['MAPPED_ID'].fillna(-1, inplace=True)
base_isaac['MAPPED_ID'] = base_isaac['MAPPED_ID'].astype(str).str.split(',', expand=False)
base_isaac = base_isaac.explode('MAPPED_ID')
base_isaac['MAPPED_ID'] = base_isaac['MAPPED_ID'].astype(int)
base_isaac.set_index('MAPPED_ID', inplace=True)
base_isaac.sort_index(ascending=False)
```

```python
pohai = pd.read_excel('Pohai.xlsx', sheet_name='Sheet1')['ID']
base_pohai = pd.read_csv('../Exported/HW/nyt3.csv').drop(columns=['Unnamed: 0'])
```

```python
base_pohai['MAPPED_ID'] = pohai
base_pohai = base_pohai.query('MAPPED_ID != "PROBLEM"').reset_index(drop=True)
base_pohai['MAPPED_ID'].fillna(-1, inplace=True)
base_pohai['MAPPED_ID'] = base_pohai['MAPPED_ID'].astype(str).str.split(',', expand=False)
base_pohai = base_pohai.explode('MAPPED_ID')
base_pohai['MAPPED_ID'] = base_pohai['MAPPED_ID'].astype(int)
base_pohai.set_index('MAPPED_ID', inplace=True)
base_pohai.sort_index(ascending=False)
```

```python
all_data = pd.concat([base_mccord, base_isaac, base_pohai])
all_data.sort_index(ascending=False)
```

```python
comparison_set = pd.read_csv('../Exported/HW/Master.csv').drop(columns=['Unnamed: 0'])
comparison_set.set_index('***ID***', inplace=True)
comparison_set.head()
```

```python
comparison_set.info()
```

```python
# Okay, now time for the fun part
# We need to create a master set that has all of the data from the three sets, and then we need to compare it to the master set
# There will be a column 'is_match' that will be 1 if the data matches, and 0 if it doesn't

training_set = pd.DataFrame(columns=['is_match', 'title_a', 'author_a', 'publisher_a', 'publish_year_a', 'isbn_a', 'title_b', 'author_b', 'publisher_b', 'publish_year_b', 'isbn_b'])
```

```python
# Iterate through the data
import sys
all_data_len = int(len(all_data))
comparison_len = int(len(comparison_set))
counter = 0
expected_matches = 0
for index, row in all_data.iterrows():
    counter += 1
    comp_counter = 0
    if int(index) == -1:
        # We know there is no match for this entry in the master set
        # For now, we'll just leave it out of the training set.
        continue
    else:
        expected_matches += 1
    print(f'{(counter / all_data_len) : 0.00%} ({counter}/{all_data_len})', end='\r', flush=True)
    data_set = comparison_set.copy()
    data_set['is_match'] = 0
    data_set.loc[int(index), 'is_match'] = 1
    data_set['title_b'] = row['title']
    data_set['author_b'] = row['authors']
    data_set['publisher_b'] = row['publisher']
    data_set['publish_year_b'] = row['publish_year']
    data_set['isbn_b'] = row['isbn10']
    data_set.columns = ['isbn_a', 'title_a', 'author_a', 'publisher_a', 'publish_year_a', 'REMOVED', 'is_match', 'title_b', 'author_b', 'publisher_b', 'publish_year_b', 'isbn_b']
    data_set = data_set[['is_match', 'title_a', 'author_a', 'publisher_a', 'publish_year_a', 'isbn_a', 'title_b', 'author_b', 'publisher_b', 'publish_year_b', 'isbn_b']]
    training_set = pd.concat([training_set, data_set])

print(f'Completed {counter * comparison_len} permutations. Expected {expected_matches} matches.')
training_set.head(5)
```

```python
training_set.sort_values(by=['is_match'], ascending=False, inplace=True)
```

```python
# We probably don't want all of the not-matches, so let's just take a random sample of them.
matches = training_set.query('is_match == 1')
print(f'Number of matches: {len(matches)}')
```

```python
not_matches = training_set.query('is_match == 0').sample(n=400000, random_state=42)
print(f'Number of not-matches: {len(not_matches)}')
```

```python
training_set = pd.concat([matches, not_matches])
training_set.to_csv('../training_set.csv', index=False)
```
