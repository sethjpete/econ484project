```python
df2 = pd.read_csv('GoodReads/books.csv', index_col=0)
df2.head(3)
```

```python
# Merged Data Sets
book_set = pd.merge(df1, df2, on='isbn', how='inner')
book_set
```

```python
# Import bestseller data
bestsellers = pd.read_csv('NYTBest/bestsellers.csv')
bestsellers = bestsellers[['isbn10', 'rank', 'price', 'weeks_on_list']]
bestsellers.head(3)
```

```python
bestsellers = bestsellers.sort_values(by='weeks_on_list', ascending=False)
bestsellers.drop_duplicates(subset='isbn10', keep='first', inplace=True)
bestsellers = bestsellers[pd.to_numeric(bestsellers['isbn10'], errors='coerce').notnull()]
bestsellers
```

```python
book_set = pd.merge(book_set, bestsellers, on='isbn10', how='left')
book_set['rank'] = book_set['rank'].fillna(-1)
book_set['price'] = book_set['price'].fillna(-1)
book_set['weeks_on_list'] = book_set['weeks_on_list'].fillna(-1)
book_set.head(3)
```

```python
book_set.columns
```

```python
# More data cleaning
def min_length(x, y):
    if len(str(x)) < len(str(y)):
        return x
    else:
        return y
    
book_set['title'] = book_set.apply(lambda x: min_length(x['title_x'], x['title_y']), axis=1)
book_set['authors'] = book_set.apply(lambda x: min_length(x['authors_x'], x['authors_y']), axis=1)
book_set['avg_rating'] = book_set.apply(lambda x: statistics.mean([x['average_rating_x'], x['average_rating_y']]), axis=1)
book_set['num_pages'] = book_set.apply(lambda x: statistics.mean([x['num_pages_x'], x['num_pages_y']]), axis=1)
book_set['num_ratings'] = book_set.apply(lambda x: statistics.mean([x['ratings_count_x'], x['ratings_count_y']]), axis=1)
book_set['best_seller'] = book_set['rank'].apply(lambda x: 1 if x > 0 else 0)
subset = ['isbn10', 'title', 'subtitle', 'authors', 'categories', 'description', 'avg_rating', 'num_pages', 'num_ratings', 'best_seller', 'rank', 'price', 'weeks_on_list', 'publisher', 'publication_date']
book_set = book_set[subset]
book_set.sort_values(by = 'weeks_on_list').head(3)
```

```python
# Generate checkout data for each book by year
prefix = 'Checkouts_By_Title_Data_Lens_' # prefix for file names
suffix = '.csv' # suffix for file names
years = [2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
for year in years:
    df = pd.read_csv('SPL/' + prefix + str(year) + suffix)
    df = df[['BibNumber']]
    df['Checkouts'] = 1
    df = df.groupby('BibNumber').sum().reset_index()
    df.columns = ['BibNum', 'Checkouts_' + str(year)]
    book_set = pd.merge(book_set, df, on='BibNum', how='left')
```

```python
book_set['subtitle'].fillna('None', inplace=True)
book_set['description'].fillna('None', inplace=True)
book_set['categories'].fillna('None', inplace=True)
book_set['Subjects'].fillna('None', inplace=True)
for year in years:
    book_set['Checkouts_' + str(year)].fillna(0, inplace=True)
book_set.sort_values(by='Checkouts_2017', ascending=False).head(3)
```
